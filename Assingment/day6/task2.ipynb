```python
from PIL import Image
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def convert_image(image_file_path):
    ifp = os.path.normpath(image_file_path.strip('""'))
    pil_img = Image.open(ifp)
    img_array = np.array(pil_img)
    return img_array
```

```python
path = input("Image path: ")
arr = convert_image(path)
```

```python
def arr_to_img(image_array):
    image = Image.fromarray(image_array)
    
    plt.imshow(image)
    plt.axis('off')  
    plt.show()
```

```python
arr_to_img(arr)
```
